```python
import matplotlib.pyplot as plt
```

```python
log10ps = []
for i in range(1, 300):
    line_count = 0
    for line in open("./data/remove_sim/CF_risk_remove{}_AFF.regenie".format(i)):
        line_count += 1
        if line_count > 1:
            parsed_line = line.strip().split()
            log10p = parsed_line[11]
    log10ps.append(float(log10p))

```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6), dpi=120)
ax.plot(log10ps, color='red')
ax.hlines(7.30, 0, 300, color='black', linestyle='--', label='Genome-wide threshold p=5e-8')
ax.legend()
ax.set_xlabel('Number of carriers of CF-risk variant in control randomly removed')
ax.set_ylabel('Burden test significance -log10(p)')
plt.show()
```
